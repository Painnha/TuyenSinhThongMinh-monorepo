```python
import pandas as pd
```

```python
df = pd.read_csv('universities.csv')
df.head()
```

```python
# In ra đầy đủ các cột của các ngành học thuộc phương thức xét tuyển thi THPT của trường IUH, trùng tên ngành vẫn in ra
nganh_hoc_iuh = df[(df['Type'].str.contains('Điểm chuẩn theo phương thức Điểm thi THPT năm 2024')) & (df['SchoolID'].str.contains('IUH'))][['Major', 'SubjectGroup', 'Benchmark_2024', 'Note']].values.tolist()
print("Danh sách các ngành học thuộc phương thức xét tuyển thi THPT của trường IUH:")
for stt, (nganh_hoc, mon_hoc, diem_chuan, ghi_chu) in enumerate(nganh_hoc_iuh, start=1):
    print(f"{stt}. {nganh_hoc} - {mon_hoc} - Điểm chuẩn: {diem_chuan} - Ghi chú: {ghi_chu}")
```
